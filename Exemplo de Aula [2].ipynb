# Exemplo de Aula [2]

### 1. Importanto as bibliotecas

```python
%matplotlib inline # usado para plotar os gráficos
import matplotlib.pyplot as plt # biblioteca gráfica
import numpy as np # biblioteca matemática
from scipy.integrate import odeint # bolioteca para integração de EDOs
```

### 2. Definindo as funções

```python
def u(t): # entrada
    return 2 # valor da vazão a partir do instante t=0

# Equação diferencial 
def model(y,t):
    return ((u(t)-0) - 4*(y-1))/5.0

def analitica(t):
  return 0.5*(1.0 - np.exp(-0.8*t))
```

### 3. Integrando e resovendo o modelo analítico

```python
IC = [1] # condição inicial de y
t = np.linspace(0,20,101) # 0 até 20 minutos (usando 101 pontos)
yr = odeint(model,IC,t) # integrando a EDO (valor inicial)
ya = analitica(t) # solução analítica
```

### 4. Plotando os resultados

```python
plt.plot(t,yr-1,'o',t,ya) # plotando os resultados em termos de variável desvio
plt.xlabel('tempo')
plt.ylabel('y (desvio)')
plt.title('Variação de y')
plt.legend(['Solução Numérica','Solução Analítica']);
plt.grid()
```

```python
plt.plot(t,yr,'o',t,ya+1) # plotando os resultados em termos de variável "original"
plt.xlabel('tempo')
plt.ylabel('y')
plt.title('Variação de y')
plt.legend(['Solução Numérica','Solução Analítica']);
plt.grid()
```
